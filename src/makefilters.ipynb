```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nside=256
k_min=3
k_max_suggestion=110
step=7

numFilts=k_max_suggestion//step-k_min//step+1
print("numFilts=",numFilts)
print("k_max=",str(k_min+(numFilts-1)*step))

filters=np.zeros((numFilts,nside,nside,nside//2+1))
karr=np.fft.fftfreq(nside,1/nside)
X,Y,Z=np.meshgrid(karr,karr,karr[:nside//2+1],indexing="ij")
k_abs=np.sqrt(X**2+Y**2+Z**2)
for i in range(numFilts):
    temp=1-np.abs(k_abs-i*step-k_min)/step
    filters[numFilts-1-i,:,:,:]+=np.where(temp<0,0,temp)
```

```python
def get_fracdevanderr(bk,bkerr,bkfid,bkfiderr):
    scale=np.abs(bk/bkfid)
    frac=np.sqrt(bkerr**2/bk**2+bkfiderr**2/bkfid**2)
    return bk/bkfid,scale*frac
```

```python
def get_order2scatim(dat,numFilts):
    im=np.zeros((numFilts,numFilts+1))
    im[:,0]=dat[0]
    c=0
    for i in range(numFilts):
        im[i,1]=dat[c]
        c+=1
        for j in range(numFilts):
            if j<=i:
                continue
            im[i,j+1-i]+=dat[c]
            c+=1
    return im
```

```python
def order2scatim(dat,numFilts):
    im=np.zeros((numFilts,numFilts+1))
    im[:,0]=dat[0]
    c=0
    for i in range(numFilts):
        im[i,1]=dat[c]
        c+=1
        for j in range(numFilts):
            if j<=i:
                continue
            im[i,j+1-i]+=dat[c]
            c+=1
    plt.imshow(im.T)
    plt.hlines(0.5,-0.5,numFilts-0.5,color="red")
    plt.hlines(1.5,-0.5,numFilts-0.5,color="red")
    return im
```
